```python
import pandas as pd
import numpy as np
#Loading the data
df= pd.read_csv('Summer.csv')
```

# Filtering numerical data

```python
df_filtered=df.query('price < 50 & rating > 4')
df_filtered
```

# Filtering categorical data

```python
df_filtered=df.query('currency_buyer == "EUR" & theme =="summer"')
df_filtered
```

# Filtering based on data contains

```python
df.loc[df['title_orig'].str.contains('Women')]
# To obtain all data which do not contain
# data.loc[~data['title_orig'].str.contains('Women')]
```

# Filtering based on regexp

```python
import re
df.loc[df['title_orig'].str.contains('women|new',regex=True,flags=re.I)]
```

# Regexp starts with

```python
df.loc[df['product_id'].str.contains('^5e',regex=True,flags=re.I)]
```
